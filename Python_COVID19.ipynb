```python
import json
# librairies pour la visualisation
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

# dfinitions des variables ncessaires (listes)
# mags : ampleurs des tremblements de terre
# lons : longitudes du tremblement de terre
# lats : latitude du tremblement de terre
# hover_texts : titre du tremblement de terre
confirmed_list = []
lons = []
lats = []
hover_texts = []

# Lisons le contenu ENTIER du fichier JSON
with open('csvjson.json','r') as file:
    all_data = json.loads(file.read())
    
for x in all_data:
    confirmed = x['3/25/20']
    lon = x['Long']
    lat = x['Lat']
    title = x['Country/Region']
    
    confirmed_list.append(confirmed)
    lons.append(lon)
    lats.append(lat)
    hover_texts.append(title)
    
all_data
```

```python
# Prparation du graphique
graph_data = [{
    'type': 'scattergeo',
    'lon': lons,
    'lat': lats,
    'text': hover_texts,
    'marker': {
        #'size': [4*mag for mag in mags], # la taille des bulles
        'color': confirmed_list,
        'colorscale': 'Viridis', # autre options 'RdBu', 'Cividis'
        'colorbar': {'title': 'Ampleur'},
    },
}]
# Titre du graphique
graph_layout = Layout(title='COVID-19')

# Gnration du graphique, comme fichier extern 'covid.html'
fig = {'data': graph_data, 'layout': graph_layout}

offline.plot(fig, filename='covid.html')
```
